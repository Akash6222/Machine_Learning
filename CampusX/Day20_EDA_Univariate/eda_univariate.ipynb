```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Categorical Data

a) Countplot

```python
sns.countplot(x=df['Survived'])                # this will give bar plot
# df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x=df['Pclass'])
```

b) Piechart

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Numerical Data

a) Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=90)
```

b) Displot

```python
sns.distplot(df["Age"],bins=90)     # y axis is probability density function
```

```python
# displot gives skewness of our data  https://miro.medium.com/v2/resize:fit:891/1*7U_8hKJyts77ZtcsBvDtmQ.png
# We can also use pandas funcyion to give skewness
df['Age'].skew()
```

c) Boxplot

```python
# Boxplot can be used to find outliers 
# see here https://cdn1.byjus.com/wp-content/uploads/2020/10/Box-Plot-and-Whisker-Plot-1.png
```

```python
sns.boxplot(x=df['Age'])
```

```python
sns.boxplot(x=df['Fare'])
```
